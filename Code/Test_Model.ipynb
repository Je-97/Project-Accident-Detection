```python
# import necessary libraries
import os
import cv2
from ultralytics import YOLO
```

# Load Best Model

```python
best_model = YOLO('/content/best.pt')
```

# Images Testing

```python
# for testing image
# Define the input image and output directory
input_image = '/content/traffic_img129_jpg.rf.5f2b657ee5bd2f6eede14c7f853554de.jpg'
output_dir = '/content/output/'

os.makedirs(output_dir, exist_ok=True)

# Read image
image = cv2.imread(input_image)

# Apply the model to the image
results = best_model(image)

# Draw bounding boxes on the image
annotated_image = results[0].plot()

# the output path for the annotated image
output_image = os.path.join(output_dir, 'car2_out.jpg')

# Save the image with detection
cv2.imwrite(output_image, annotated_image)
```

# Open Video

```python
video_path = '/content/video1.mp4'
cap = cv2.VideoCapture(video_path)
print(cap.isOpened()) # To insure the video is opened
```

```python
# Get Video information
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
```

```python
# Writer Video
output_video = 'output_acc2.mp4'
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video, fourcc, fps, (width, height))
```

# Testing Video

```python
# Here is about read the video frame by frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Here is about applying the model to the frame
    results = best_model(frame)

    # Here is about draw bounding box
    annotated_frame = results[0].plot()

    # Here is about write  to output
    out.write(annotated_frame)

# Here is about release and close videos
cap.release()
out.release()
cv2.destroyAllWindows()
```
